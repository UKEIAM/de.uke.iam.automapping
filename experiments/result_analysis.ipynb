```python
import pandas as pd
```

##### Ground truth

```python
df_evaluation=pd.read_excel('source_to_concept_map_new.xlsx')
```

```python
df_evaluation['prepeared_source_code']=df_evaluation['prepeared_source_code'].str.lower()
```

##### Results from scikit learn

```python
df_mine=pd.read_csv('result_2.csv', index_col = [0])
```

```python
df_mine=df_mine.astype({"Concept_id": "Int64"})
```

```python
df_mine
```

##### Merge results

```python
df_ground_scikit=pd.merge(df_evaluation, df_mine, how='left', left_on='prepeared_source_code', right_on='Term')
```

```python
df_ground_scikit=df_ground_scikit[['source_code', 'source_code_description', 'target_concept_id', 'Concept', 'Term', 'Concept_id', 'Similarity']]
```

```python
df_ground_scikit=df_ground_scikit.drop_duplicates().rename(columns={"Concept": "concept_scikit", "Concept_id": "concept_id_scikit", 'Term':'term_scikit'})
```

```python
df_ground_scikit
```

```python
import numpy as np
df_ground_scikit.index=np.arange(len(df_ground_scikit))
```

```python
df_ground_scikit=df_ground_scikit.fillna(0)
```

```python
n=0
for i in range(len(df_ground_scikit)):
    if df_ground_scikit['target_concept_id'][i]==df_ground_scikit['concept_id_scikit'][i]:
        n=n+1
```

```python
per=n/len(df_ground_scikit)*100
```

```python
per
```

```python
df_ground_scikit
```

```python
df_ground_scikit.to_excel('ground_scikit.xlsx', index=False)
```

##### Data from Usagi 

```python
df_usagi=pd.read_csv('concept_evaluation_new.csv', index_col = [0])
```

```python
df_usagi['sourceName']=df_usagi['sourceName'].str.lower()
```

```python
df_usagi=df_usagi[['sourceName', 'matchScore', 'targetConceptId', 'targetConceptName']]
```

##### Merge usagi and Scikit

```python
df_usagi_scikit=pd.merge(df_usagi, df_mine, how='left', left_on='sourceName', right_on='Term')
```

```python
df_usagi_scikit=df_usagi_scikit.drop_duplicates().rename(columns={"sourceName": "usagi_term", "matchScore": "usagi_similarity_score", 'targetConceptId':'usagi_concept_id', "targetConceptName":"usagi_concept", "Term":'scikit_term', 'Concept':'scikit_concept', 'Concept_id':'scikit_concept_id', 'Similarity':'scikit_similarity_score'})
```

```python
df_usagi_scikit.index=np.arange(len(df_usagi_scikit))
```

```python
df_usagi_scikit=df_usagi_scikit.fillna(0)
```

```python
m=0
for i in range(len(df_usagi_scikit)):
    if df_usagi_scikit['usagi_concept_id'][i]==df_usagi_scikit['scikit_concept_id'][i]:
        m=m+1
```

```python
percent=m/len(df_usagi_scikit)*100
```

```python
percent
```

```python
df_usagi_scikit.to_excel('usagi_scikit.xlsx', index=False)
```
